```python
import pandas as pd
```

```python
# Create a data frame with details of 10 students and columns as Roll Number, Name, Gender, Marks1, Marks2, Marks3
data_frame={
    'Roll Number':[2101,2102,2103,2104,2105,2106,2107,2108,2109,2110],
    'Name':["Aarav","Ananya","Arjun","Aadhya","Advik","Vihaan","Ishaan","Aarohi","Rohan","Aditya"],
    'Gender':["Male","Female","Male","Female","Male","Male","Male","Female","Male", "Male"],
    'Marks1':[78, 45, 92, 63, 87, 54, 33, 73, 68, 81],
    'Marks2':[49, 88, 93, 26, 81, 57, 90, 62, 81, 41],
    'Marks3':[83, 91, 75, 42, 89, 68, 95, 53, 86, 34]
}
```

```python
df = pd.DataFrame(data_frame)
df
```

```python
#Create a new column with total marks
df['Total Marks']=df['Marks1']+df['Marks2']+df['Marks3']
df
```

```python
#Find the lowest marks in Marks1
lowest_marks1=df['Marks1'].min()
print("The lowest marks in Marks1 is ", lowest_marks1)
```

```python
#Find the Highest marks in Marks2
highest_marks2=df['Marks2'].max()
print("The highest marks in Marks2 is ", highest_marks2)
```

```python
# Find the average marks in Marks3
average_marks3=df['Marks3'].mean()
print("The average marks of Marks3 is ", average_marks3)
```

```python
#Find student name with highest average
student_highest_avg=df.loc[df['Total Marks'].idxmax(), 'Name']
print('The student with highest average marks is', student_highest_avg)
```

```python
#Find how many students failed in Marks2 (<40)
marks2_fail=df[df['Marks2'] < 40].shape[0]
print('The number of students failed are ' , marks2_fail)
```
